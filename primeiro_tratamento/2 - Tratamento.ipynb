### Concatenando as tabelas: Rental + Inventory + Film_category + Category + film

```python
rental.head()
```

```python
inventory.head()
```

```python
# Concatenando as tabelas "rental" e "inventory" através da coluna 'inventory_id'
alugueis = pd.merge(rental, inventory, on=['inventory_id'], how='inner')

# Visualizando as 5 primeiras linhas do novo DF
alugueis.head()
```

##### Concatenando as tabelas "alugueis" + "film_category" através da coluna 'film_id'

```python
film_category.head()
```

```python
# Concatenando as tabelas 
alugueis = pd.merge(alugueis, film_category, on=['film_id'], how='inner')

# Exibir a tabela completa resultante (alugueis)
alugueis.head()
```

```python
category.head()
```

```python
# Concatenando as tabelas "alugueis" + "category" através da coluna 'category_id', para puxar o nome da categoria
alugueis = pd.merge(alugueis, category, on=['category_id'], how='inner')

# Exibir a tabela completa resultante (alugueis)
alugueis.head()
```

```python
# Concatenando as tabelas "alugueis" + "film" através da coluna film_id, para acrescentar as colunas 'rental_duration, rental_rate, replacement_cost'
alugueis = pd.merge(alugueis, film[['film_id', 'rental_duration', 'rental_rate', 'replacement_cost']], on='film_id', suffixes=('_alugueis', '_film'))

# Exibir a tabela completa resultante (alugueis)
alugueis.head()
```

```python
# Verificando se há valores nulos (NaN) ou em branco em cada coluna
valores_nulos = alugueis.isnull().sum()
print("Quantidade de valores nulos (NaN) ou em branco em cada coluna:")
print(valores_nulos)
```

```python
print(alugueis.dtypes)
```

### Concatenando as tabelas "alugueis" + "customer"


```python
# Concatenando first_name + last_name
customer['name_customer'] = customer['first_name'].str.cat(customer['last_name'], sep=' ')

# Exibir os resultados
customer.head()
```

```python
# Selecionando apenas a coluna "name_customer" do DataFrame "customer"
customer_names = customer[['customer_id', 'name_customer']]

#Concatenando as tabelas 
df1 = pd.merge(alugueis, customer_names, on='customer_id')

# Exibir a tabela completa resultante (df1)
df1.head()
```

```python
# Verificando se há valores nulos (NaN) ou em branco em cada coluna
valores_nulos = df1.isnull().sum()
print("Quantidade de valores nulos (NaN) ou em branco em cada coluna:")
print(valores_nulos)
```

```python
# Concatenando as tabelas
customer_completa = pd.merge(customer, address_country, on='address_id')

# Exibir a tabela completa resultante (customer_completa)
customer_completa.head()
```

```python
#Verificando as duplicadas 
duplicados = customer_completa.duplicated(subset='customer_id')
print("Total de linhas", len(customer_completa))

#Contando o número total de linhas duplicadas através do método 'sum()':
num_duplicados = duplicados.sum()
print("Número de linhas duplicadas:", num_duplicados)
```
