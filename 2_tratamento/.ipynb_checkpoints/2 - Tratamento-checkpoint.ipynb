# Tratamentos para Análise -  EMPRESA DISCORAMA

## Importações

```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
```

### Leitura CSV

```python
actor = pd.read_csv("actor.csv")
address = pd.read_csv("address.csv")
category = pd.read_csv("category.csv")
city = pd.read_csv("city.csv")
country = pd.read_csv("country.csv")
customer= pd.read_csv("customer.csv")
film = pd.read_csv("film.csv")
film_actor = pd.read_csv("film_actor.csv")
film_category = pd.read_csv("film_category.csv")
inventory = pd.read_csv("inventory.csv")
payment = pd.read_csv("payment.csv")
rental = pd.read_csv("rental.csv")
staff = pd.read_csv("staff.csv")
store = pd.read_csv("store.csv")
```

### Excluindo as colunas desnecessárias para as análises

```python
# Atribuindo um novo DataFrame sem a coluna "last_update" 
actor = actor.drop(columns='last_update')
address = address.drop(columns='last_update')
city = city.drop(columns='last_update')
country = country.drop(columns='last_update')
category = category.drop(columns='last_update')
customer = customer.drop(columns='last_update')
film = film.drop(columns='last_update')
rental = rental.drop(columns='last_update')
staff = staff.drop(columns='last_update')
store = store.drop(columns='last_update')
```

```python
film_category.head()
```

### Concatenando Staff + Store

```python
# Concatenando as tabelas
staff_store = pd.merge(staff, store, on=['store_id'], how='inner')

# Exibir a tabela completa resultante (staff_store)
staff_store.head()
```

```python
# Dropando colunas desnecessárias para análise
staff_store = staff_store.drop(columns=['password', 'picture'])
```

```python
#Renomeando a coluna 
staff_store.rename(columns={'address_id_x': 'address_id'}, inplace=True)
```

```python
# Exibir a tabela completa resultante (staff_store)
staff_store.head()
```

### Concatenando Rental + Inventory + Film_category + Category + film

```python
# Concatenando as tabelas
alugueis = pd.merge(rental, inventory, on=['inventory_id'], how='inner')

# Exibir a tabela completa resultante (alugueis)
alugueis.head()
```

```python
# Dropando colunas desnecessárias para análise
alugueis = alugueis.drop(columns=['last_update'])
```

```python
# Concatenando as tabelas
alugueis = pd.merge(alugueis, film_category, on=['film_id'], how='inner')

# Exibir a tabela completa resultante (alugueis)
alugueis.head()
```

```python
# Concatenando as tabelas
alugueis = pd.merge(alugueis, category, on=['category_id'], how='inner')

# Exibir a tabela completa resultante (alugueis)
alugueis.head()
```

```python
# Concatenando as tabelas
alugueis = pd.merge(alugueis, film[['film_id', 'rental_duration', 'rental_rate', 'replacement_cost']], on='film_id', suffixes=('_alugueis', '_film'))

# Exibir a tabela completa resultante (alugueis)
alugueis.head()
```

```python
print(alugueis.dtypes)
```

### Concatenando alugueis + customer


```python
# Concatenando first_name + last_name
customer['name_customer'] = customer['first_name'].str.cat(customer['last_name'], sep=' ')

# Exibir os resultados
customer.head()
```

```python
# Selecionando apenas a coluna "name_customer" do DataFrame "customer"
customer_names = customer[['customer_id', 'name_customer']]

#Concatenando as tabelas 
df1 = pd.merge(alugueis, customer_names, on='customer_id')

# Exibir a tabela completa resultante (df1)
df1.head()
```

#### Concatenando Customer + Address + City + Country

```python
# Concatenando as tabelas
address_completa = pd.merge(address, city, on=['city_id'], how='inner')

# Exibir a tabela completa resultante (address_completa)
address_completa.head()
```

```python
print(address_completa.dtypes)
```

```python
print(country.dtypes)
```

```python
# Concatenando as tabelas
address_country = pd.merge(address_completa, country, on='country_id')

# Exibir a tabela completa resultante (address_country)
address_country.head()
```

```python
print(address_country.dtypes)
```

```python
# Concatenando as tabelas
customer_completa = pd.merge(customer, address_country, on='address_id')

# Exibir a tabela completa resultante (customer_completa)
customer_completa.head()
```

```python
#Verificando as duplicadas 
duplicados = customer_completa.duplicated(subset='customer_id')
print("Total de linhas", len(customer_completa))

#Contando o número total de linhas duplicadas através do método 'sum()':
num_duplicados = duplicados.sum()
print("Número de linhas duplicadas:", num_duplicados)
print("-=" *18)

# Linhas duplicadas no DataFrame:
linhas_duplicadas = customer_completa[duplicados]
print("Linhas duplicadas:")
print(linhas_duplicadas)
```
